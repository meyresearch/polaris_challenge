```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("leaderboards/antiviral-potency-2025/aggregated.csv")
```

```python
data
```

```python
data.columns
```

# Kendall tau

```python
columns = ['kendall_tau']#, 'mean_absolute_error', 'mean_squared_error', 'pearsonr', 'r2', 'spearmanr']
highlight_name = ['caithmac','hbaumann','ialibay']

for column in columns:
    c_data = data[column].values
    test = [c.split("±") for c in c_data]
    test = np.array(test).astype('float')
    df = pd.DataFrame()
    df[column] = test[:,0]
    df['error']= test[:,1]
    df['owner_slug'] = data['owner_slug'].values
    sorted_df = df.sort_values(by=[column], ascending=False)
    plt.rcParams["figure.figsize"] = [7.50, 3.50]
    colors = ['blue' if name in highlight_name else 'red' for name in sorted_df['owner_slug'].values]
    sns.barplot(sorted_df, x='owner_slug', y=column, hue = colors, **{'yerr':df['error']})
    plt.xticks(rotation=90)


```
